```python
class Human:
  def __init__(self, name, age):
        self.name = name
        self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です")
    else:
      print(f"{self.name}は大人ではありません")

# Humanクラスのインスタンスを複数生成してリストに追加
humans = [
    Human("侍太郎", 36),
    Human("侍花子", 26),
    Human("侍若君", 18)
]
# リストの各要素に対してcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```

```python
from PIL import Image
import numpy as np

# 画像を読み込む
im = Image.open("camera.jpg")
im = im.resize((im.width //2, im.height //2)) # 元の画像はデカいので縮小
im

```

```python
# PIL形式からNumPy形式に変換
im_np = np.asarray(im)
im_np
```
