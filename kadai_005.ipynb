```python
import matplotlib.pyplot as plt
import numpy as np

# 
# 010100
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# 
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

plt.show()
```
