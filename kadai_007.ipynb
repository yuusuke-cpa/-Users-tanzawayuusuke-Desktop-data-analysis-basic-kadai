```python
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

# CSVファイルを読み込む
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')

# 内部結合でDataFrameをマージする
df1 = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# カテゴリーごとの個数を集計し、棒グラフをプロットする
category_counts = df1['カテゴリー'].value_counts()
category_counts.plot(kind='bar')
plt.show()

# 集計関数（sum）
grouped_sum = df1.groupby('商品番号')['注文数'].sum()

# 基本統計量の算出
grouped_describe = df1.groupby('商品番号')['注文数'].describe()

# 結果の出力
print(grouped_sum)
print(grouped_describe)
```
